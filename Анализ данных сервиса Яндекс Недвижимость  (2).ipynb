### Предобработка данных

#### Работа с пропусками

```python
#Посмотрим общее количество пропусков по столбцам
print(data.isna().sum())
```

```python
#приведем столбец cityCenters_nearest в соответствие
data.rename(columns={'cityCenters_nearest' : 'city_centers_nearest'}, inplace=True)
```

```python
# Изучим столбец is_apartment
data['is_apartment'].unique()
```

```python
#Возможно, данный стобец появился не сразу либо сознательно не заполняли владельцы квартир, поэтому заменим пропуски на False
data['is_apartment'] = data['is_apartment'].fillna(False) 
```

```python
#Пропуски в столбце балкон заменим на 0, предполагая, что в данных обьектах недвижимости их нет.
data['balcony'] = data['balcony'].fillna(0)
```

```python
#проверим гипотезу, что пропуски  в столбцах parks_around3000,parks_around3000 совпадают по строчкам. Скорее всего, пропуски  сделаны в населенных пунктах, где нет ни прудов, ни парков.
data.loc[(data['parks_around3000'].isna() == True) & (data['ponds_around3000'].isna() == True)]
```

```python
#заменим пропуски  в ponds_around3000 и parks_around3000 на 0
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
```

```python
data['parks_around3000'] = data['parks_around3000'].fillna(0)
```

```python
# удалим строки с пропущенными названиями населенных пунктов. Их не так много, а восстановить их проблематично.
data = data.dropna(subset=['locality_name']) 
```

```python
#удалим этажность
data = data.dropna(subset=['floors_total'])  
```

```python
#предположим, то пропуски в days_exposition свидетельствуют о том, что обьявления на момент формирования таблицы еще не были проданы. Поэтому пометим их специальным значением, например 9999
data['days_exposition'] = data['days_exposition'].fillna(9999)
```

```python
#Посмотрим на пропуски в столбце 'ceiling_height'
data['ceiling_height'].unique()
```

```python
#в ceiling_height  заменим пропуски на медианным значением по общей площади
ceiling_height_median = data['ceiling_height'].median()
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_height_median)
```

```python
data['ceiling_height'].isna().sum()
```

```python
#в living_area  заменим пропуски на медианным значением по жилой площади
for i in data['total_area'].unique():
    data.loc[(data['living_area'].isna()) | (data['total_area'] == i),'living_area'] = data.loc[data['total_area'] == i,'living_area'].median()
```

```python
data['living_area'].isna().sum()
```

```python
# заменим пропущенные значения площади кухни  в квартирах-студиях на нули

data.loc[data['studio'] == True, 'kitchen_area']  = data.loc[data['studio'] == True, 'kitchen_area'] == 0
```

```python
data[data['studio']]
```

```python
#в kithen_area остальные пропуски заменим на медианным значением по жилой площади
for i in data['living_area'].unique():
    data.loc[(data['kitchen_area'].isna()) | (data['living_area'] == i),'kitchen_area'] = data.loc[data['living_area'] == i,'kitchen_area'].median()
```

```python
data['kitchen_area'].isna().sum()
```

```python
#проверим все ли пропуски, которые мы хотели, обработаны
print(data.isna().sum())
```

Вывод: Для нашего исследования не все пропуски требовали заполнения, например, ближайшие парки и пруды в радиусе 3 км мы не трогали.
Часть пропусков возникла из-за того,что, например, не имея рядом пруда или парка, программа не проставляла 0, а  оставляла незаполненным данное поле. Так же, как и с числом балконов. Возможно, некоторые поля нужно автоматизировать таким образом, чтобы не было возможности не заполнять определенные поля. 
Кухонная площадь, в большинстве случаев, могла оставаться незаполненный из-за принадлежности к свободной планировки, а жилая проставлялась, например в общей площади.

```python
# вынесем в отдельный столбец тип населенного пункта
def type_of_locality (locality_name):
    if 'поселок' in locality_name:
        return 'поселок'
    elif 'поселок городского типа' in locality_name:
        return 'пгт'
    elif 'поселок' in locality_name:
        return 'поселок'
    elif'садовое'  in locality_name:
        return 'садовое товарищество'
    elif 'село' in locality_name:
        return 'село'
    elif 'деревня' in locality_name:
        return 'деревня'  
    return 'нет названия'
     
```

### Посчитайте и добавьте в таблицу новые столбцы

```python
# посчитаем цену одного квадратного метра 
data['price_of_meter'] = data['last_price']/data['total_area']
```

```python
#добавим день недели публикации объявления (0 — понедельник, 1 — вторник и так далее); 
data['day_of_exposition'] = data['first_day_exposition'].dt.weekday
```

```python
#добавим месяц публикации объявления;
data['month_of_exposition'] = data['first_day_exposition'].dt.month
```

```python
#добавим год публикации объявления;
data['year_of_exposition'] = data['first_day_exposition'].dt.year
```

```python
#добавим тип этажа квартиры (значения — «‎первый», «последний», «другой»);
# создайте функцию floor_type(row)
def floor_type (row):
    floor = row['floor']
    floors_total = row['floors_total']
    
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'
```

```python
# примените функцию методом apply()
data['floor_type'] = data.apply(floor_type, axis=1) 
```

```python
#расстояние до центра города в километрах (переведите из м в км и округлите до целых значений).
data['distance to city center'] =(data['city_centers_nearest']/1000).round()
```
